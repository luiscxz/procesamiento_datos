# Navegar por los datos perdidos con multiple imputaci√≥n

### *Qu√© es esto?*
**La imputaci√≥n m√∫ltiple (IM)** es una t√©cnica estad√≠stica dise√±ada para abordar el problema de los datos que faltan en un conjunto de datos. Se trata de una herramienta sofisticada en la caja de herramientas de un cient√≠fico de datos üîß, dise√±ada espec√≠ficamente para manejar las incertidumbres asociadas a la informaci√≥n que falta.

### *Qu√© hace?* 
El proceso de IM implica la creaci√≥n de m√∫ltiples versiones del conjunto de datos original, cada una de las cuales difiere ligeramente debido a los valores imputados (o rellenados) de los datos que faltan. El resultado son varios conjuntos de datos completos, cada uno de los cuales ofrece una perspectiva ligeramente distinta de los datos.

<div style="text-align: center;">
    <img src="../inputacion.png" alt="boxplot" width="900"/>
</div>

A continuaci√≥n, cada uno de estos conjuntos de datos completos se somete al an√°lisis previsto. Dado que cada conjunto de datos tiene valores imputados diferentes, los resultados del an√°lisis diferir√°n ligeramente en cada uno de ellos. Esta variaci√≥n en los resultados de los distintos conjuntos de datos proporciona una indicaci√≥n de la incertidumbre asociada a los valores que faltan.

Por √∫ltimo, el IM combina los resultados de cada conjunto de datos en un √∫nico resultado final. Este resultado final tiene en cuenta la variabilidad de los resultados de cada uno de los conjuntos de datos imputados, lo que proporciona un resultado m√°s s√≥lido y fiable que los m√©todos tradicionales de imputaci√≥n √∫nica.

Esencialmente, el IM proporciona un m√©todo completo y estad√≠sticamente s√≥lido para tratar los datos que faltan, ofreciendo una mejora significativa sobre m√©todos m√°s simples que pueden introducir sesgos o imprecisiones.

# Por qu√© es importante la imputaci√≥n?

- **Incompatible con la mayor√≠a de las librer√≠as Python usadas en Machine Learning:** S√≠, has le√≠do bien. Al utilizar las bibliotecas para ML (el m√°s com√∫n es skLearn), no tienen una disposici√≥n para manejar autom√°ticamente estos datos que faltan y puede conducir a errores.
- **Distorsi√≥n en el conjunto de datos:** Una gran cantidad de datos perdidos puede causar distorsiones en la distribuci√≥n de las variables, es decir, puede aumentar o disminuir el valor de una categor√≠a particular en el conjunto de datos.
- **Afecta al modelo final:** los datos que faltan pueden causar un sesgo en el conjunto de datos y dar lugar a un an√°lisis err√≥neo del modelo.

# An√°lisis completo de casos (CCA)

Se trata de un m√©todo bastante sencillo para tratar los datos que faltan, eliminando directamente las filas que tienen datos faltantes, es decir, s√≥lo tenemos en cuenta las filas en las que tenemos datos completos, es decir, en las que no faltan datos. Este m√©todo tambi√©n se conoce popularmente como *"Eliminaci√≥n por listas"*.
### Suposiciones:
- *Datos perdidos al azar (MAR)*
- *Los datos que faltan se eliminan completamente de la tabla*
### Ventajas:
- *F√°cil de implementar*
- *No es necesario manipular los datos*
### Limitaciones:
- *Los datos borrados pueden ser informactivos*
- *Puede llevar a la eliminaci√≥n de gran parte de los datos*
- *Puede crear un sesgo en el conjunto de datos, si se elimina de √©l una gran cantidad de un determinado tipo de variable.*
- *El modelo de producci√≥n no sabr√° qu√© hacer con los datos que faltan.*
### Cuando usar:
- *Los datos son MAR (faltan de manera aleatoria)*
- *Adecuado para datos mixtos, num√©ricos y categ√≥ricos*
- *Los datos que faltan no superan el 5% - 6% del conjunto de datos*
- *Los datos no contienen mucha informaci√≥n y no sesgar√°n el conjunto de datos.*

# Imputaci√≥n por valor arbitrario

Esta es una t√©cnica importante utilizada en Imputaci√≥n ya que puede manejar tanto las variables Num√©ricas como las Categ√≥ricas. Esta t√©cnica consiste en agrupar los valores que faltan en una columna y asignarles un nuevo valor que est√© lejos del rango de esa columna. Normalmente se utilizan valores como 99999999 o -9999999 o "Falta" o "No definido" para variables num√©ricas y categ√≥ricas.
### Suposiciones
- *Los datos no faltan de forma aleatoria*
- *Los datos que faltan se imputan con un valor arbitrario que no forma parte del conjunto de datos o Media/Media/Modo de datos*
### Ventajas
- *F√°cil de implementar*
- *Podemos utilizarlo en producci√≥n*
- *Conserva la importancia de los "valores perdidos" si existen*
### Desventajas
- *Puede distorsionar la distribuci√≥n original de las variables*
- *Los valores arbitrarios pueden crear valores at√≠picos*
- *Se requiere precauci√≥n adicional al seleccionar el valor Arbitrario*
### Cuando usar
- *Cuando los datos no son MAR (Missing At Random)*
- *Apto para todos*

# Imputaci√≥n de categor√≠as frecuentes

Esta t√©cnica consiste en sustituir el valor que falta por la variable de mayor frecuencia o, en palabras sencillas, sustituir los valores por el moda de esa columna. Esta t√©cnica tambi√©n se denomina imputaci√≥n de moda.

### Suposiciones
- *Faltan datos al azar*
- *Existe una alta probabilidad de que los datos que faltan se parezcan a la mayor√≠a de los datos*
### Ventajas
- *Facil de implementar*
- *Podemos obtener un conjunto de datos completo en muy poco tiempo*
- *Podemos utilizar esta t√©cnica en el modelo de producci√≥n*
### Desventajas
- *Cuanto mayor sea el porcentaje de valores que faltan, mayor ser√° la distorsi√≥n*
- *Puede dar lugar a una representaci√≥n excesiva de una categor√≠a concreta*
- *Puede distorsionar la distribuci√≥n original de las variables*
### Cuando usar
-  *Datos faltantes al azar(MAR)*
-  *Los datos que faltan no superan el 5% - 6% del conjunto de datos*

# Imputaci√≥n estad√≠stica de valores

Para ello se utilizan valores estad√≠sticos como la media o la mediana. Sin embargo, ninguno de ellos garantiza datos insesgados, especialmente si faltan muchos valores.

La media es m√°s √∫til cuando los datos originales no est√°n sesgados, mientras que la **mediana es m√°s robusta, no es sensible a los valores at√≠picos** y, por tanto, se utiliza cuando los datos est√°n sesgados.

En unos datos distribuidos normalmente, se pueden obtener todos los valores que est√°n dentro de 2 desviaciones est√°ndar de la media. A continuaci√≥n, rellene los valores que faltan generando n√∫meros aleatorios entre $(\text{mean} - 2 \cdot \text{std}) \, \& \, (\text{mean} + 2 \cdot \text{std})$


# Imputaci√≥n por regresi√≥n lineal

A partir de los datos existentes, se puede calcular la l√≠nea de mejor ajuste entre dos variables, a partir de la comprobaci√≥n de las correlaciones.

Cabe mencionar que los modelos de regresi√≥n lineal son sensibles a los valores at√≠picos. Y solo se debe hacer imputaci√≥n por regresi√≥n lineal si hay correlaci√≥n entre dos caracter√≠sticas.

# Imputaci√≥n iterativa (Iterative imputation) 

La imputaci√≥n iterativa se refiere a un proceso en el que cada caracter√≠stica se modela como una funci√≥n de las dem√°s caracter√≠sticas, por ejemplo, un problema de regresi√≥n en el que se predicen valores perdidos. Cada caracter√≠stica se imputa secuencialmente, una tras otra, lo que permite utilizar valores imputados anteriores como parte de un modelo para predecir caracter√≠sticas posteriores.

Es iterativo porque este proceso se repite varias veces, lo que permite calcular estimaciones cada vez mejores de los valores que faltan a medida que se estiman los valores que faltan en todas las caracter√≠sticas.

Este enfoque puede denominarse generalmente como **especificaci√≥n totalmente condicional (FCS)** o **imputaci√≥n multivariante por ecuaciones encadenadas (MICE)**.

# Imputaci√≥n por Nearest neighbors

La clase KNNImputer proporciona imputaci√≥n para rellenar los valores que faltan utilizando el enfoque k-Nearest Neighbors. Por defecto, se utiliza una m√©trica de distancia euclidiana que admite valores perdidos, nan_euclidean_distances, para encontrar los vecinos m√°s cercanos. Cada caracter√≠stica que falta se imputa utilizando los valores de n_vecinos m√°s cercanos que tienen un valor para la caracter√≠stica. Las caracter√≠sticas de los vecinos se promedian uniformemente o se ponderan por la distancia a cada vecino. Si a una muestra le falta m√°s de una caracter√≠stica, los vecinos de esa muestra pueden ser diferentes en funci√≥n de la caracter√≠stica concreta que se impute. Cuando el n√∫mero de vecinos disponibles es inferior a n_vecinos y no hay distancias definidas al conjunto de entrenamiento, se utiliza la media del conjunto de entrenamiento para esa caracter√≠stica durante la imputaci√≥n. Si hay al menos un vecino con una distancia definida, durante la imputaci√≥n se utilizar√° la media ponderada o no ponderada de los vecinos restantes. Si una caracter√≠stica falta siempre en el entrenamiento, se elimina durante la transformaci√≥n. 

# Marking imputed values

El transformador MissingIndicator es √∫til para transformar un conjunto de datos en la correspondiente matriz binaria que indica la presencia de valores perdidos en el conjunto de datos. Esta transformaci√≥n es √∫til junto con la imputaci√≥n. Cuando se utiliza la imputaci√≥n, conservar la informaci√≥n sobre los valores que han faltado puede ser informativo. Tenga en cuenta que tanto SimpleImputer como IterativeImputer tienen el par√°metro booleano add_indicator (False por defecto) que cuando se establece en True proporciona una forma conveniente de apilar la salida del transformador MissingIndicator con la salida de la imputaci√≥n.

NaN se utiliza normalmente como marcador de posici√≥n para valores perdidos. Sin embargo, obliga a que el tipo de datos sea float. El par√°metro missing_values permite especificar otro marcador de posici√≥n, como un entero. 

# Comentarios

Se puede adoptar el enfoque aleatorio, en el que rellenamos el valor que falta con un valor aleatorio. Llevando este enfoque un paso m√°s all√°, se puede dividir primero el conjunto de datos en dos grupos (estratos), bas√°ndose en alguna caracter√≠stica, por ejemplo el sexo, y luego rellenar los valores que faltan para los distintos sexos por separado, de forma aleatoria.

En la imputaci√≥n secuencial en caliente, la columna que contiene los valores que faltan se ordena seg√∫n la(s) variable(s) auxiliar(es), de modo que los registros que tienen auxiliares similares aparecen secuencialmente. A continuaci√≥n, cada valor perdido se rellena con el valor del primer registro siguiente disponible.

Lo m√°s interesante es que tambi√©n se puede utilizar la imputaci√≥n ùëò del vecino m√°s cercano, que clasifica los registros similares y los pone juntos. Un valor perdido se rellena encontrando primero los ùëò registros m√°s cercanos al registro con valores perdidos. A continuaci√≥n, se elige un valor entre los ùëò vecinos m√°s cercanos (o se calcula a partir de ellos). En el caso de la computaci√≥n, pueden utilizarse m√©todos estad√≠sticos como la media (como se ha comentado antes).

# 1. Ejemplos de imputaci√≥n de datos

A continuaci√≥n vamos a mostrar un ejemplo de imputaci√≥n de datos en el dataframe de precios de refrigeradores.

```python
# importando librer√≠as necesarias
import os
import pandas as pd
pd.set_option('future.no_silent_downcasting', True)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# accediendo a la ruta de los datos
os.chdir('D:\\15. Notas de clase\\implementaci√≥n de modelo ml para regresion lineal')
# leyendo archivo "column_2C_weka.csv"
df = pd.read_csv('fridge_price_predictor_dataset_real_brands.csv')
df.head(5)
```

Podemos observar que hay varias columnas con informaci√≥n faltante. Con el fin de simplificar este proceso, vamos a identificar los valores √∫nicos de cada categoria.

```python
# consultando valores √∫nicos
unicos = [(df[i].unique(),i) for i in df.columns]
unicos = [(i, len(unique)) for unique, i in unicos]
unicos = pd.DataFrame(unicos, columns=['Columna', 'Cantidad de valores √∫nicos'])
unicos.head(len(unicos))
```

Podemos ver que exiten columnas con muchos datos √∫nicos, como es el caso de *Model*, *Price*, *Random Numeric Feature 1* y *Random Text Feature 2*. Estas caracter√≠sticas las vamos a eliminar ya que no las consideramos necesarias y identificaremos las columnas con datos faltantes.

```python
# vamos a crear una copia del dataframe y vamos a eliminar las columnas que tiene muchos datos √∫nicos
data = df.copy()
data = data.drop(columns=['Model','Price','Random Numeric Feature 1','Random Text Feature 2'])
# veamos que columns tienen datos faltantes
print('Las columnas con datos faltantes son:', data.columns[data.isnull().any()])
```

Como se puede ver, tenemos varias columnas con datos faltantes, la inquietud es la siguiente:

1) **¬øEstos datos faltantes estan al azar, es decir, son MAR?**
2) **¬øQu√© porcerntaje de datos faltanes hay en cada columna?**

Para dar respuesta a estas inquietudes, vamos a realizar una inspecci√≥n gr√°fica de los datos faltantes y una prueba Chi-Cuadrado para evaluar si los datos faltantes est√°n al azar, es decir, si siguen un patr√≥n de "Missing At Random" (MAR).

**Las hip√≥tesis para esta prueba son las siguientes:**

**Hip√≥tesis nula (H‚ÇÄ):** La distribuci√≥n de los valores nulos en la columna con datos faltantes es independiente de las categor√≠as de la columna categ√≥rica.

**Hip√≥tesis alternativa (H‚ÇÅ):** La distribuci√≥n de los valores nulos en la columna con datos faltantes depende de las categor√≠as de la columna categ√≥rica.

**Interpretaci√≥n:** 

- **Si el P-value > 0.05**, no se rechaza la hip√≥tesis nula (H‚ÇÄ), lo que significa que no hay suficiente evidencia para concluir que los valores nulos en la columna con datos faltantes est√©n asociados con las categor√≠as de la columna categ√≥rica. En este caso, podr√≠as considerar que los datos faltantes son independientes de esa variable categ√≥rica.
- **Si el P-value < 0.05**, se rechaza la hip√≥tesis nula, lo que sugiere que s√≠ hay una relaci√≥n significativa entre los valores nulos y la columna categ√≥rica, es decir, **los datos faltantes no son al azar (MAR) respecto a esa columna categ√≥rica.**


```python
from scipy.stats import chi2_contingency
# Seleccionando columnas categ√≥ricas
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
# seleccionando conlumnas don datos faltantes
columns_with_nan = data.columns[data.isnull().any()]
# creando variable que almacena los resultados de la prueba
results = []

# Itera sobre columnas categ√≥ricas y columnas con datos faltantes
for column in categorical_columns:
    if data[column].notna().sum() > 0:  # Salta si la columna categ√≥rica est√° vac√≠a
        for col_with_nan in columns_with_nan:
            # Tabla de contingencia
            cont_table = pd.crosstab(data[col_with_nan].isnull(), data[column])
            
            # Aplica prueba Chi-Cuadrado si la tabla tiene m√°s de 1 fila y columna
            if cont_table.shape[0] > 1 and cont_table.shape[1] > 1:
                chi2, p, dof, _ = chi2_contingency(cont_table)
                
                # Almacena resultados en lista
                results.append({
                    'Column_with_NaN': col_with_nan,
                    'Categorical_Column': column,
                    'Chi2': chi2,
                    'P-value': p,
                    'Degrees_of_Freedom': dof
                })

# Crea DataFrame con resultados
results_df = pd.DataFrame(results)
```

# Imputaci√≥n MICE con bosques aleatorios

Dado que la imputaci√≥n por XGboost no es adecuada, procedemos a implementar bosques aleatorios. En este caso se generar√°n **cinco dataset** completamente imputados, para posteriormente combinar y promediar las imputaciones.

```python
import miceforest as mf

df1_dummy = datadummy.copy()
impute = mf.ImputationKernel(
    df1_dummy,
    datasets=5, #  Indica que se generar√°n 5 datasets completos imputados
    save_all_iterations=True, # Guarda todas las iteraciones realizadas durante el proceso de imputaci√≥n
    random_state=42
)
```

```python
# realizando imputaci√≥n con 5 iteraciones
impute.mice(5)
# obteniendo los resultados
impute.complete_data(0)
# combinando y promediando las imputaciones. Se redondea a entero ya que las imputaciones en este caso se espera que sean enteros
df1_dummy_imputed = pd.concat([impute.complete_data(i) for i in range(5)]).groupby(level=0).mean().round(0).astype(int)
```

```python
# veamos los resultados de esta imputaci√≥n
plt.figure(figsize=(6,3), dpi=200)
sns.kdeplot(data = data['Features'], label='Datos sin imputar', fill = True)
sns.kdeplot(data = df1_dummy_imputed['Features'], label='Datos imputados', fill = True)
plt.title('Gr√°fico de densidad de datos inputados y sin imputar ')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
```

# Imputaci√≥n mediante red neuronal MIDASpy

MIDASpy es una librer√≠a en Python que utiliza una red neural de autoencoder variacional para la imputaci√≥n de datos faltantes. Est√° dise√±ada espec√≠ficamente para manejar datos tabulares con valores faltantes en m√∫ltiples columnas. 

Veamos como es su aplicaci√≥n.

```python
import MIDASpy as md
# creamos una copia del dataframe
data_midas = df.copy()
# vamos a eliminar las siguientes columnas
list_del_col = ['Model','Price','Random Text Feature 2']# eliminando columnas
data_midas= data_midas.drop(columns=list_del_col )
# seleccionando nombres de columnas categoricas
categorical = data_midas.select_dtypes('object').columns.to_list()
```

Procedemos a las variables categoricas y las codificaremos usando la funci√≥n de preprocesamiento cat_conv incorporada en MIDASpy, que devuelve tanto los datos codificados como una lista anidada de nombres de columnas categ√≥ricas que podemos pasar al algoritmo de imputaci√≥n. 

```python
# codificando variables categ√≥ricas
data_cat, cat_cols_list = md.cat_conv(data_midas[categorical])
# veamos las variables codificadas
data_cat.head()
```

```python
# veamos los nombres de las columnas que se crearon en la codificaci√≥n
print(cat_cols_list)
```

Una vez realizado este paso, es necesario eliminar del dataframe, las columnas catrgoricas utilizadas para realizar la codificaci√≥n.

```python
# eliminando variables categ√≥ricas del dataframe
data_midas.drop(categorical, axis = 1, inplace = True)
data_midas.head()
```

En este paso, tenemos el dataframe sin las columnas categ√≥ricas. Ahora lo que vamos a hacer es, concatenar este dataframe con las columnas codificadas.

```python
# concatenando variables categ√≥ricas codificadas
constructor_list = [data_midas]
constructor_list.append(data_cat)
data_midas_concat = pd.concat(constructor_list, axis=1)
data_midas_concat.head()
```

Una vez el dataframe esta con todas las columnas categoricas codificadas, procedemos a buscar los datos faltantes en el dataframe y los reemplazamos con **np.nan** ya que es el formato de datos faltantes admito por la red neuronal.

```python
# procedemos a buscar datos faltantes en el nuevo dataframe
nan_loc = data_midas_concat.isnull()
# vamos a reemplazar estos faltantes con np.nan ya que as√≠ lo requiere la red neuronal
data_midas_concat[nan_loc] = np.nan
# vamos a crear el modelo
```

Con los datoa ya preparados, es momento de definir nuestro modelo, en este caso vamos a crear el modelo con las siguientes caracter√≠sticas.

- **Dos capas ocultas con 356 neuronas cada una**
- **Variational Autoencoder (VAE)**, Se usar√° como **False**
- **Se fijar√° una semilla**
- **Se apagar√°n el 75% de las neuronas en cada √©poca, con el fin de evitar el **overfitting (sobreajuste)**

```python
# vamos a definir el modelo
imputer = md.Midas(layer_structure = [356,356], # el modelo tendr√° dos capas ocultas con 356 neuronas
                   vae_layer = False, # Este par√°metro indica que no se utilizar√° un VAE (Variational Autoencoder), sino que se usar√° un autoencoder cl√°sico.
                   seed = 89, # Fija una semilla para asegurar la reproducibilidad.
                   input_drop = 0.75 # dropout
                   )
```

una vez definido el modelo, procedemos a construirlo y entrenarlo. 

```python
# contruyendo el modelo
imputer.build_model(data_midas_concat, softmax_columns = cat_cols_list)
# entrenando el modelo durante 20 √©pocas
imputer.train_model(training_epochs = 20)
```

Una vez finalizada todas las √©pocas de entrenamiento, procedemos a generar 10 dataframes imputados.

```python
imputations = imputer.generate_samples(m=10).output_list 
```

Ahora vamos a obtener los dataframes imputados con las columnas originales.

```python
flat_cats = [cat for variable in cat_cols_list for cat in variable]
# Para cada dataframe imputado:
# Iterar sobre cada dataframe imputado
for i in range(len(imputations)):
    # Obtener los valores originales de las categor√≠as basados en idxmax
    tmp_cat = [imputations[i][x].idxmax(axis=1) for x in cat_cols_list]
    
    # Crear un DataFrame con las variables categ√≥ricas originales
    cat_df = pd.DataFrame({categorical[j]: tmp_cat[j] for j in range(len(categorical))})
    
    # Reemplazar las columnas codificadas por las categor√≠as originales
    imputations[i] = pd.concat([imputations[i], cat_df], axis=1).drop(flat_cats, axis=1)
    
    # Eliminar los prefijos de las columnas categ√≥ricas
    for col in categorical:
        imputations[i][col] = imputations[i][col].str.replace(f'{col}_', '', regex=False)
    
    # Redondear y convertir las columnas num√©ricas a enteros ya que nuestras catorias se codificar√°n en enteros positivos
    imputations[i] = imputations[i].round(0)
```
